```python
def getsize(file_name,tally,sa):
    genome_seq = ''
    with open(file_name,'r') as genome_file:
        genome_file.readline()
        genome_seq = genome_file.read().replace('\n', '');
    
    size = len(genome_seq)+6*(4+1)
    if tally == 0:
        size += len(genome_seq)*4*5
    else:
        size += len(genome_seq)*4*5/tally
    if sa == 0:
        size += 2*4*len(genome_seq)
    else:
        size += 2*4*len(genome_seq)/sa
    return size/1024/1024

def func(file_name):
    spaces = []
    for tally in [8,32,128,512]:
        for sa in [4,16,64,254]:
            spaces.append(getsize(file_name,tally,sa))
    return pd.DataFrame({'space [MB]':spaces})
            
    

display(getsize('coffea_arabica.fa',0,0))
display(getsize('mus_pahari.fa',0,0))
display(getsize('e_coli.fa',0,0))
```

```python
data = pd.read_csv('performances/coffea_arabica.csv')
term1 = data[0:16]

term2 = data[16:32]['time']
term3 = data[32:48]['time']
space_table = term1[['tally_factor','sa_factor']]

space = func('coffea_arabica.fa')

space_table=pd.merge(space_table, space, right_index=True,left_index=True, how='inner')


display(space_table)





data = pd.read_csv('performances/mus_pahari.csv')
term1 = data[0:16]

term2 = data[16:32]['time']
term3 = data[32:48]['time']
space_table = term1[['tally_factor','sa_factor']]

space = func('mus_pahari.fa')

space_table=pd.merge(space_table, space, right_index=True,left_index=True, how='inner')


display(space_table)



data = pd.read_csv('performances/e_coli.csv')
term1 = data[0:16]

term2 = data[16:32]['time']
term3 = data[32:48]['time']
space_table = term1[['tally_factor','sa_factor']]

space = func('e_coli.fa')

space_table=pd.merge(space_table, space, right_index=True,left_index=True, how='inner')


display(space_table)
```
